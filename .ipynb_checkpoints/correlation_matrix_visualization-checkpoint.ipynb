```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
```

```python
root = Path('aggregated_data')
joinData_file = 'aggregate_data.csv'
joinData_df = pd.read_csv(root / joinData_file)
```

```python
plt.figure(figsize=(8,8))
plt.matshow(joinData_df.corr(),fignum=1)
plt.xticks(range(len(joinData_df.corr().columns)), labels=joinData_df.corr().columns, rotation='vertical')
plt.yticks(range(len(joinData_df.corr().columns)), labels=joinData_df.corr().columns)

plt.imshow(joinData_df.corr(), cmap=plt.cm.get_cmap('YlGnBu', 360))
plt.colorbar()
plt.clim(-1, 1);
plt.show()
```
